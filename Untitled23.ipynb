```python
import matplotlib.pyplot as plt
#sample data
students=['A','B','C','D','E']
marks=[85,70,90,60,75]
```

```python
import matplotlib.pyplot as plt
#bar chart
plt.bar(students,marks,color="skyblue")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("student performance")
plt.show()
```

```python
#scatter plot
plt.scatter(students,marks,color="green")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("student performance")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[85,70,90,60,75]
plt.plot(x,y,marker="o")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("student performance")
plt.show()
```
