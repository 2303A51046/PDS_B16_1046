```python
import matplotlib.pyplot as plt

subjects = ['Math', 'Science', 'English', 'History']

avg_marks = [78, 72, 85, 67]

bars = plt.bar(subjects, avg_marks, color='skyblue', edgecolor='black')

for bar in bars:

    height = bar.get_height()

    xpos = bar.get_x() + bar.get_width() / 2

    plt.text(xpos, height + 1, f'{height}', ha='center', va='bottom')

plt.xlabel('Subjects')

plt.ylabel('Average Marks')

plt.title('Average Marks of Students in Different Subjects')

plt.ylim(0, 100)

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()


```
