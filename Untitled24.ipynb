```python
import matplotlib.pyplot as plt
subjects=["Math","Physics","chemistry","Biology","English"]
marks=[85,90,78,88,78]
```

```python
plt.plot(subjects,marks,marker="o",color="green")
plt.title("student marks in different subjects")
plt.xlabel("subjects")
plt.ylabel("marks")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
days=["mon","tue","wed","thur","fri","sat","sun"]
temperature=[30,32,31,29,33,35,54]
```

```python
plt.plot(days,temperature,marker="o",linestyle="--",color="red")
plt.title("temperature in a week")
plt.xlabel("days")
plt.ylabel("temperature")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
months=["jan","feb","mar","apr","may","jun"]
sale=[15000,18000,12000,22000,24000,20000]
plt.bar(months,sale,color="green")
plt.title("monthly sales report")
plt.xlabel("months")
plt.ylabel("sale in usd")
plt.grid(axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
days=["mon","tue","wed","thur","fri","sat","sun"]
stock_price=[150,152,148,155,160,156,162]
plt.plot(days,stock_price,marker="^",linestyle="-",color="red",linewidth=2)
plt.title("stock price movement")
plt.xlabel("days")
plt.ylabel("stock price in usd")
plt.grid(True)
plt.show()
```
